# 0. 라이브러리 임포트

```python
import numpy
import pandas
```

# 1. 2022/12/29 일자의 KOSPI 종목 정보

```python
df_kospi_info = pandas.read_csv('data-0101-info.csv', dtype={'Code': object})
df_kospi_info = df_kospi_info[['Code', 'Name', 'Marcap']]
df_kospi_info['종목코드'] = 'A' + df_kospi_info['Code']
df_kospi_info = df_kospi_info[['종목코드', 'Name', 'Marcap']]
df_kospi_info.columns = ['종목코드', '종목명', '시가총액']
df_kospi_info = df_kospi_info.set_index(['종목코드'])
display(df_kospi_info)
print(df_kospi_info.head(30))
```

# 2. KOSPI 지수 정보

```python
df_kospi_index = pandas.read_csv('data-0102-kospi.csv')
df_kospi_index = df_kospi_index[['eod_date', 'price_close']]
df_kospi_index.columns = ['기준일자', 'KOSPI지수-종가']
df_kospi_index = df_kospi_index.set_index(['기준일자'])
display(df_kospi_index)
```

# 3. 가격/거래 정보

```python
df_prices = pandas.read_csv('data-0103-prices.csv.gz', compression='zip')
df_prices.rename(columns={'eod_date': '기준일자'}, inplace=True)
df_prices = df_prices.set_index(['기준일자'])
display(df_prices)
```

# 4. 상위 30개 종목 검토

### 4.1. 상위 30개 종목 데이터 시가/종가 추출

```python
list_top_30 = df_kospi_info.head(30).index.tolist()

set_cols = set(list(df_prices.columns))
list_new = list()
for equity_code in list_top_30:
    if equity_code + '-price_open' in set_cols:
        list_new.append(equity_code + '-price_open')
    if equity_code + '-price_close' in set_cols:
        list_new.append(equity_code + '-price_close')
df_top30 = df_prices[list_new]
display(df_top30)
```

### 4.2. 데이터 시작지점 확인

```python
list_start = list()
for col_name in df_top30.columns:
    if 'price_open' in col_name:
        continue
    df_part = df_top30[col_name]
    df_part = df_part.dropna()
    list_start.append({
        '종목코드': col_name.replace('-price_close', ''),
        '시작일자': numpy.min(df_part.index),
    })
df_check_start = pandas.DataFrame(list_start)
df_check_start = df_check_start.set_index(['종목코드'])
df_kospi_info_days = df_check_start.join(df_kospi_info, how='left')
display(df_kospi_info_days)
```

### 4.3. 2001/01/02 부터 데이터가 존재하는 종목만 선택

```python
df_kospi_info_days = df_kospi_info_days[df_kospi_info_days['시작일자'] == 20010102]
print(df_kospi_info_days.shape)
display(df_kospi_info_days)
```

### 4.4. 삼성전자우 와 삼성전자는 동일회사이므로 제외

```python
df_kospi_info_days = df_kospi_info_days[df_kospi_info_days['종목명'] != '삼성전자우']
print(df_kospi_info_days.shape)
display(df_kospi_info_days)
```

### 4.5. 카카오는 테스트 기간 중 우회 상장으로인해 기업가치가 연속적이지 않으므로 제외

```python
df_kospi_info_days = df_kospi_info_days[df_kospi_info_days['종목명'] != '카카오']
print(df_kospi_info_days.shape)
display(df_kospi_info_days)
```

### 4.6. 최종적으로 상위 10개 종목 선택

```python
df_kospi_info_days = df_kospi_info_days.head(10)
print(df_kospi_info_days.shape)
display(df_kospi_info_days)
```

# 5. 선택 종목과 코스피지수 가격 정보 설정

### 5.1. 코스피지수 설정

```python
df_base = df_kospi_index.copy()
print(df_base.shape)
display(df_base)
```

### 5.2. 선택 종목에 대한 시가/종가 정보 설정

```python
df_info = df_kospi_info_days[['종목명']]
display(df_info)

list_cols = list()
list_cols_new = list()
for equity_code, row in df_info.iterrows():
    list_cols.append(equity_code + '-price_open')
    list_cols.append(equity_code + '-price_close')
    list_cols_new.append(equity_code + '-' + row['종목명'] + '-시가')
    list_cols_new.append(equity_code + '-' + row['종목명'] + '-종가')

df_selected = df_prices[list_cols]
df_selected.columns = list_cols_new
display(df_selected)
```

### 5.3. 최종 기초 데이터 생성

```python
df_base = df_base.join(df_selected, how='left')
print(df_base.shape)
display(df_base)
```

```python
df_base.to_csv('data-0201-base.csv')
```
